```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
sns.set()
```

Plotting a Vector

```python
plt.quiver(0,0,4,5)
plt.show()
```

```python
plt.quiver(0,0,4,5, scale_units='xy', angles='xy', scale=1)
plt.xlim(-8, 8)
plt.ylim(-8,8)
plt.show()
```

```python
plt.quiver(0,0,4,5, scale_units='xy', angles='xy', scale=1, color='b')
plt.xlim(-8, 8)
plt.ylim(-8,8)
plt.show()
```

```python
plt.quiver(0,0,4,5, scale_units='xy', angles='xy', scale=1, color='b')
plt.quiver(0,0,-2,-5, scale_units='xy', angles='xy', scale=1, color='y')
plt.xlim(-8, 8)
plt.ylim(-8,8)
plt.show()
```

Addition of Two Vectors

```python
vector_1 = np.asarray([0,0,2,3])
vector_2 = np.asarray([0,0,3,-2])
sum = vector_1 + vector_2
print(sum)
```

```python
plt.quiver(0,0,2,3, scale_units='xy', angles='xy', scale=1, color='b')
plt.quiver(0,0,3,-2, scale_units='xy', angles='xy', scale=1, color='y')
plt.quiver(0,0,5,1, scale_units='xy', angles='xy', scale=1, color='r') #sum of above vector
plt.xlim(-8, 8)
plt.ylim(-8,8)
plt.show()
```

Subtraction of 2 Vectors

```python
vector_1 = np.asarray([0,0,2,3])
vector_2 = np.asarray([0,0,3,-2])
differance = vector_1 - vector_2
print(differance)
```

```python
plt.quiver(0,0,2,3, scale_units='xy', angles='xy', scale=1, color='b')
plt.quiver(0,0,3,-2, scale_units='xy', angles='xy', scale=1, color='y')
plt.quiver(0,0,-1,5, scale_units='xy', angles='xy', scale=1, color='r') #Subtraction of above vector
plt.xlim(-8, 8)
plt.ylim(-8,8)
plt.show()
```

Multiplying a vector by a scalar

```python
vector_1 = np.asarray([0,0,2,3])

vector_2 = 2 * vector_1
vector_3 = -0.5 * vector_1
print(vector_2)
print(vector_3)
```

```python
plt.quiver(0,0,2,3, scale_units='xy', angles='xy', scale=1, color='b')
plt.quiver(0,0,4,6, scale_units='xy', angles='xy', scale=1, color='y') #multiply value of scalar
plt.quiver(0,0,-1,-1.5, scale_units='xy', angles='xy', scale=1, color='r')
plt.xlim(-8, 8)
plt.ylim(-8,8)
plt.show()
```
