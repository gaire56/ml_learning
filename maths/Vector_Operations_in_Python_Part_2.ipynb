```python
import numpy as np
```

Dot Product of 2 Vectors

image.png

```python
a = np.array([2, 3])

b = np.array([4, 4])

a_dot_b = np.dot(a, b)

print(a_dot_b)
```

```python
c = np.array([13, 16, 27])

d = np.array([14, 72, 54])

c_dot_d = np.dot(c, d)

print(c_dot_d)
```

Cross Product of two Vectors

cross product of Vectors.png

```python
a = np.array([2, 3])

b = np.array([4, 4])

a_cross_b = np.cross(a, b)

print(a_cross_b)
```

```python
c = np.array([13, 16, 27])

d = np.array([14, 72, 54])

c_cross_d = np.cross(c, d)

print(c_cross_d)
```

Projection of 'a' vector on 'v' vector

image.png

```python
a = np.array([2, 5])

v = np.array([4, -3])

#magnitude of "v" vector
magnitude_of_v = np.sqrt(sum(v**2))

proj_of_a_on_v = (np.dot(a, v)/magnitude_of_v**2)*v

print('Projection of a vector on v vector = ', proj_of_a_on_v)
```

```python
a = np.array([13, 16, 27])

v = np.array([14, 72, 54])

#magnitude of "v" vector
magnitude_of_v = np.sqrt(sum(v**2))

proj_of_a_on_v = (np.dot(a, v)/magnitude_of_v**2)*v

print('Projection of a vector on v vector = ', proj_of_a_on_v)
```
